```python
#Drop the columns not needed for the analysis and store it into another variable titled 'f1_comp'
f1_comp = compiled.drop(['What it does', 'Investor', 'Founders', 'Unnamed: 9', 'About Company'],axis=1)

"""In total, we have 7 columns and 2868 rows to be used for analysis and inference making
some of these rows contain null values which would have to be strategically eliminated or negelcted 
so as to have clear insight from our data"""
```

```python
#Replace all the '—' values with zero(0)
f1_comp['Amount'] = f1_comp['Amount'].astype(str).replace('—', '0', regex=True)


#Replace all the '' values with zero(0)
f1_comp['Amount'] = f1_comp['Amount'].astype(str).replace('', '0', regex=True)
#f1_comp = f1_comp.replace('', '0', regex=True)


# Replace all the 'NaN' & 'nan' values with zero(0)

f1_comp['Amount'] = f1_comp['Amount'].astype(str).replace(['NaN', 'nan'], '0', regex=True)
#f1_comp = f1_comp.replace(['NaN', 'nan'], '', regex=True)

#f1_comp = f1_comp.replace('—', '0', regex=True)
```

# Convert the Amount Column which is in 'str' to 'int32'
>In processing the data, i noticed that i could not directly process the amount coulumn from object to integer. When ever I do, it returned an error(indefinate values cannot be converted). Thus, I had to first force it into a float, replace the blanks with zero, calculate the mean and relace the zero's with the mean, then finally convert the valuesfor the amount colun into an integer. 
>The code below was used in processing the data in that regard:

```python
#converting the amount to an integer process
f1_comp['Amount'] = pd.to_numeric(f1_comp['Amount'],errors = 'coerce')

#since its a float, the blanks would have to be replaced with zero to allow for the calculation of the mean age. the following function was used for that:
f1_comp['Amount'].replace([' '], 0, inplace=True)

#this code was used in calculating the mean value for the values in the amount column.
mn = f1_comp['Amount'].mean()
mn

#fill the zero's or missing values with the mean(mn)
f1_comp['Amount'].fillna(value=mn, inplace=True)

#Use this line of code to finally convert the values in the Amount column from float to integers
f1_comp['Amount'] = f1_comp['Amount'].astype(int)
```

# Question on: 
>The relationship between Sector of Operation and the Amount of Funding received

```python
top_10_sectors = f1_comp[['Sector', 'Amount']].groupby(['Sector'])['Amount'].size().nlargest(10).reset_index(name='top10')

top_10_sectors.plot.bar(x='Sector', y='top10', rot=0, xlabel = 'Sector of Operation', ylabel='Total Amount (Count)', title = 'Ten most funded sectors')

def customize():
    plt.rcParams['figure.figsize'] = (40,30)
    plt.rcParams.update({'font.size': 30})
    plt.xticks(rotation=90)
    plt.show()
    return
customize();
```

# Answer:
> From the chart, it is eveident that most of the startups operating tech based ventures ranked amongst those startups likely to receive funding from invetors.

> In the case of start ups operating in the product and service sector, the high number is liley due to the support services these startups provide to compliment the activities of the startups in the other sectors.

> It can be concluded that if startups operate in tech based sectors, or provide important complimentary services to other business and society in general it is likely to recieve  high amount of funding.
